# <font color='blue'> 4. 각각의 숫자를,  제곱한 값으로 바꾸세요.

```python
dictionary = {'C1' : [10,20,30],
              'C2' : [20,30,40]}
```

```python
dictionary_C1 = dictionary['C1']
dictionary_C1[0] = dictionary_C1[0]**2
dictionary_C1[1] = dictionary_C1[1]**2
dictionary_C1[2] = dictionary_C1[2]**2
```

```python
dictionary_C2 = dictionary['C2']
dictionary_C2[0] = dictionary_C2[0]**2
dictionary_C2[1] = dictionary_C2[1]**2
dictionary_C2[2] = dictionary_C2[2]**2
```

```python
dictionary
```

```python
# C1: 첫번째 요소 제곱하기
dictionary['C1'][0] = dictionary_C1[0]**2
dictionary['C1'][1] = dictionary_C1[1]**2
dictionary['C1'][2] = dictionary_C1[2]**2
# C2: 두번째 요소 제곱하기
dictionary['C2'][0] = dictionary_C2[0]**2
dictionary['C2'][1] = dictionary_C2[1]**2
dictionary['C2'][2] = dictionary_C2[2]**2
print(dictionary)
```
