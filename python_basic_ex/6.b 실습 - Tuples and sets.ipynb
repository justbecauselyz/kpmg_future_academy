# <font color='blue'> 6. my_list에는 동일한 값들이 들어있습니다. 동일한 값들을 제거하여, 유니크한 값들의 리스트로 만드세요.  </font>

```python
my_list = ["2", 50, "2", 50, 4, 6, 9, 9]
```

```python
# 동일한 값들을 제거하여
unique_val = set(my_list)
unique_val
```

```python
# 유니크한 값들의 리스트로 만드세요
list(unique_val)
```

```python
# 한 셀로 통합
unique_val = set(my_list)
list(unique_val)
```
