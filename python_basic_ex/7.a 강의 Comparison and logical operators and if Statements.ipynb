## 조건문이란, 우리 일상생활에서 흔히 쓰는 문장.  if ~ 라면, ~ 한다.

### 비가 오면 우산을 들고 나가고, 비가 오지 않으면 놓고간다.

### 5가 2보다 크면 condition is True 를 출력하고, 그렇지 않으면, condition is False를 출력한다.

```python
# 조건 + 행동(액션) ==> 조건문
```

```python
# if 조건(True|False) :
  # 행동(실행할 문장)
```

```python
my_score = 50
cut_line = 60
# 참인 경우에만 실행할 때
if my_score > cut_line:
  print("합격")
```

```python
# 참인 경우와 거짓인 경우 각각에 판단해서 실행
my_score = 50
cut_line = 60
if my_score > cut_line:
  print("합격")
else:
  print("불합격")
```

```python
# 60 미만: 더 공부하세요
# 60~80 미만: 합격했어요
# 80 이상: 최고예요
score = 90
cut_line = 60
if score < cut_line:
  print("더 공부하세요")
elif score >= 60 and score < 80 : # 또 다른 조건
  print("합격했어요")
else:
  print("최고예요")
```

```python
if False:
  print("조건이 참이면 실행")
else:
  print("조건이 거짓이면 실행")
```

## CPU의 코드 실행 순서

### 위의 식을 작성해봅시다.

```python
apple_cnt = 3
if apple_cnt >= 10:
  print("사과를 구매할 필요 없다")
elif apple_cnt >= 4:
  print("사과를 적정량 유지하고 있다")
elif apple_cnt >= 2:
  print("사과를 구매할 시기입니다")
else:
  print("사과를 구매해야 합니다")
```

## Nested 중첩된 조건문

```python
year = 2025
month = 8
day = 31
```

```python
day=day+1
if day>31:
  month=month+1
  day=1
  if month>12:
    year=year+1
    month=1
print(year, month, day)
```
