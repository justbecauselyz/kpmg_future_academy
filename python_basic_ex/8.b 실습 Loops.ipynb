or

```python
# map(각 요소에 적용할 함수, 리스트)
input_vals = "5 7 -23 -1 99 321"
vals=input_vals.split(" ")   # 공백을 기준으로 각 요소를 분리해서 리스트로 만듦
vals_casted=list(map(int, vals))   # vals의 모든 요소가 int로 형변환 됨
vals_casted.reverse()   # 역순으로 배열
print(vals_casted)
```

# <font color='blue'> 9. 유저가 점수를 입력하면, 해당 학점을 출력하는 프로그램입니다.  
# <font color='blue'> 단, 유저가 -1 을 입력하면, 프로그램을 종료합니다. -1 을 입력하기 전에는 계속 입력을 받습니다.


- 사용자가 어떤 값을 입력함. 그 값에 따라 A, B, C, F 등급을 표시함
- 숫자 <= 73 : F
- 74 <= 숫자 < 84 : C
- 84 <= 숫자 < 94 : B
- 94 <= 숫자 < 100: A



```python
# -1을 입력하기 전에는 계속 입력을 받습니다.
```

```python
while True:
  num=int(input("점수를 입력하세요.> "))
  if 0<=num<=73:
    print("F")
  elif 74<=num<84:
    print("C")
  elif 84<=num<94:
    print("B")
  elif 94<=num<=100:
    print("A")
  if num==-1:
    print("종료")
    break
```

답안:

```python
while True:
  num=int(input("점수를 입력하세요.> "))
  if num==-1:
    print("종료")
    break
  else:
    # 숫자 <= 73 : F
    if num<=73:
      print("F")
    # 74 <= 숫자 < 84 : C
    elif 74<=num<84:
      print("C")
    # 84 <= 숫자 < 94 : B
    elif 84<=num<94:
      print("B")
    # 94 <= 숫자 <= 100: A
    elif 94<=num<=100:
      print("A")
```

## 업그레이드 문제
- 문자를 입력했을 때, 예외처리(다시 입력하도록 함)
- 1~100 이외의 숫자를 입력했을 때, 예외처리(다시 입력하도록 함)

```python
while True:
  num=int(input("점수를 입력하세요.> "))
  if 0<=num<=73:
    print("F")
  elif 74<=num<84:
    print("C")
  elif 84<=num<94:
    print("B")
  elif 94<=num<100:
    print("A")
  if num==-1:
    break
  else:
    print("다시 입력하세요")
```

답안:

```python
while True:
  input_val=input("점수를 입력하세요.(1~100)> ")
  # 문자를 입력 했다면 : pass
  if input_val.isalpha():
    pass
  else:
    num=int(input_val)
    if num==-1:
      print("종료")
      break
    else:
      if num<=73:
        print("F")
      elif 74<=num<84:
        print("C")
      elif 84<=num<94:
        print("B")
      elif 94<=num<=100:
        print("A")
```

# <font color='blue'> 10. 유저한테 구구단 시작단을 입력받으세요.
# <font color='blue'> 2를 입력받으면, 2단에서 9단까지 출력하세요.
# <font color='blue'> 8을 입력받으면, 8단에서 9단까지 출력하세요.
# <font color='blue'> 음수 입력하거나 9단보다 큰 단을 입력하면, “잘못입력하였습니다.” 하고, 다시 단을 입력 받습니다.
# <font color='blue'> 단, 0 을 입력받으면, 프로그램을 종료하세요.
# <font color='blue'> 입력)  단을 입력하세요.
# <font color='blue'> 출력)  4X1 = 4
# <font color='blue'>          ...


```python
mul_table = int(input("구구단 시작단 : "))
if mul_table<=0 or mul_table>9:
  print("잘못입력하였습니다.")
else:
  for i in range(1,10):
    print(f"{mul_table} x {i} = {mul_table*i}")
```

답안:

```python
# 2를 입력받으면, 2단에서 9단까지 출력하세요.
# 8을 입력받으면, 8단에서 9단까지 출력하세요.
```

```python
num=int(input("숫자를 입력하세요.> "))
for i in range(num,10):  # 5~9단까지
  for j in range(1,10):
    print(f"{i} x {j} = {i*j}")
  print("-"*20)
```

```python
while True:
  input_val=input("0~9 숫자를 입력하세요.")
  if input_val.isalpha() or int(input_val)<0 or int(input_val)>9:
    print("잘못입력하였습니다.")
    pass
  else:
    # 0일 경우는 종료
    num=int(input_val)
    if num==0:
      break
    else:
    # 0이 아닐 경우는 곱셈 출력
      for i in range(num,10):  # 5~9단까지
        for j in range(1,10):
          print(f"{i} x {j} = {i*j}")
        print("-"*20)
```

```python
# 반복문 구현
# 음수 입력하거나 9단보다 큰 단을 입력하면, “잘못입력하였습니다.” 하고, 다시 단을 입력 받습니다.
# 단, 0 을 입력받으면, 프로그램을 종료하세요.

while True:
    input_val = input("0~9 숫자를 입력하세요.")
    if input_val.isalpha() or int(input_val) < 0 or int(input_val) > 9 :
        pass
    else:
        num = int(input_val)
        # 0일 경우 -> 종료
        if num == 0:
            print("종료")
            break
        else:
            # 0이 아니경우 -> 곱셈출력
            for i in range(num, 9+1):  # 5~ 9단까지
                for j in range(1, 9+1):
                    print(f"{i} x {j} = {i * j}")
                print("-"*10)
```
