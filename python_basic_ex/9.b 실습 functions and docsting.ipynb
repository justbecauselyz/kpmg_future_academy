# <font color='blue'> 2. 숫자 두개, 문자열 1개를 입력받아서, 더하기와 곱하기 연산을 하고, 리턴하는 함수를 만드세요.
## 함수명은 cal ( value1, value2, operator_str )
## operator_str 은 'add', 'mul' 두가지 입니다.
## 만약 cal(3, 4, 'add') 이면 더하기를 하고
## cal(3, 4, 'mul') 이면 곱하기를 합니다.

```python
def cal(value1,value2,operator_str):
  # pass
  # add
  if operator_str=="add":
    return value1+value2
    # mul
  elif operator_str=="mul":
    return value1*value2
  else:
    # 잘못입력 예외 처리
    return "입력값이 잘못됐습니다."
```

```python
# 호출부
value1=int(input("첫 번째 숫자를 입력하세요."))
value2=int(input("두 번쨰 숫자를 입력하세요."))
operator_str=input("연산자를 입력하세요.")
cal(value1,value2,operator_str)
```

# 3. 과일 등급을 반환하는 함수를 작성하세요.

## 과일 크기를 입력으로 받으면, 등급을 리턴하는 함수.
    
## 1에서10 : C등급, 11에서20 : B등급, 21에서30 : A등급
    

```python
def fruit_size():
  if 1<=fruit_size<=10:
    return "C등급"
  elif 11<=fruit_size<=20:
    return "B등급"
  elif 21<=fruit_size<=30:
    return "A등급"
```

```python
fruit_size=int(input("과일 크기를 입력하세요."))
fruit_size(f"{fruit_size}")
```

풀이:

```python
# 정의부
# 과일 크기를 입력으로 받으면, 등급을 리턴하는 함수
def fruit_grade(size):
  if 1<=size<=10:
    return "C등급"
  elif 11<=size<=20:
    return "B등급"
  elif 21<=size<=30:
    return "A등급"
  else:
    return "등급 측정 불가"
```

```python
# 호출부
# 과일의 크기를 전달
# size=25
size=int(input("과일 크기를 입력하세요."))
fruit_grade(size)
print(f"사이즈 {size}는 {fruit_grade(size)}입니다.")
```
