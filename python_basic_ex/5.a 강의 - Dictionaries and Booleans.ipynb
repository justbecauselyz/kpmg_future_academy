# DICTIONARIES

- my_dict = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
- 딕쳐너리는 키 밸류의 쌍으로 되어 있다.
- 키는, 딕셔너리 안에 유일한 값으로 되어 있다. 따라서 키가 같은 값을 가질 수 없다.  
   그러나 밸류는 같은 값이 여러개 있어도 상관없다.
- 리스트는 인덱스의 오프셋으로 접근하지만, 딕셔너리는 키로 접근한다.

## dict : Read

```python
# dic create
my_phone = {"YuJeong":"010-1234-5678", "mom":"010-5678-1234"}
my_phone
```

```python
# 데이터 엑세스(read)
my_phone['YuJeong']
```

[tip] 리스트 데이터 엑세스: create, read, update, delete

```python
# KeyError: 'color' --- IGNORE ---
my_phone['sister']
```

```python
# 없으면 None 반환 --- IGNORE ---
# None: 없는 데이터를 표현하는 데이터 타입
print(type(my_phone.get('sister')))
```

## dict : Update

```python
my_phone = {'브랜드': "Apple", '모델명': "MacBook Pro", '가격': 2000000, '색상': "Silver", 'year': 2020}
my_phone
```

```python
my_phone['가격'] = 2500000
my_phone
```

```python
# 특정 값 추가
my_phone['생산국가'] = '미국'
my_phone
```

## dict : Delete

```python
del my_phone['year']
my_phone
```

## dict 객체에서 사용할 수 있는 함수

```python
# key 추출
key_list = my_phone.keys()
key_list
```

```python
# key 추출한 값을 저장한 변수는 type이 dict_keys 임
# list형으로 쓰고 싶으면 list()로 형변환을 시켜야 함
key_list = list(key_list)
type(key_list)
```

```python
# dict의 값(value)만 추출하기
my_phone.values()
```

```python
my_phone.items()
```

```python
my_phone
```

```python
# 여러개의 key에 대해서 한꺼번에 update할 때
my_phone.update({"모델명":"아이패드", "가격":1000000, "색상":"gray"})
my_phone
```

```python
# key가 있는지 확인(있으면 True, 없으면 False)
"브랜드" in my_phone.keys()
# "브랜드" in my_phone
```

```python
"아이패드" in my_phone.values()

```

```python
"color" in my_phone
```
