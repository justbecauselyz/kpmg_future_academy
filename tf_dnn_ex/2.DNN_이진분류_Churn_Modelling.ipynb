# Neural Networks and Deep Learning

```python
# import lib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
data_path = 'datas_dnn/Churn_Modelling.csv'
df = pd.read_csv(data_path)
```

```python
df.isna().sum()
```

```python
df.head(4)
```

```python
y = df['Exited']
```

```python
X = df.loc[:,'CreditScore':'EstimatedSalary']
```

```python
y.unique()
```

```python
X.head(3)
```

```python
X['Geography'].unique()
```

```python
X['Gender'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([('onehot', OneHotEncoder(), [1]), ('label', OrdinalEncoder(), [2])], remainder='passthrough')
```

```python
X_ct = ct.fit_transform(X)
X_ct[0,:]
```

```python
X.head(3)
```

```python
X_ct
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
X_ct
```

```python
sc = MinMaxScaler()
```

```python
X_scaled = sc.fit_transform(X_ct)
X_scaled.shape
```

```python
X_scaled[0,:]
```

```python
X.head(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_scaled.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# modeling
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(keras.Input(shape=(12,)))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=8, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

```python
# compile
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy']
             )
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=10)
```

```python
X_train.shape
```

```python
model.evaluate(X_test, y_test)
```
