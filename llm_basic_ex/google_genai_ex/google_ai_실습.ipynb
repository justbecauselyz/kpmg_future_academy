```python
from dotenv import load_dotenv
import os

# 환경변수 로딩
load_dotenv(override=True)

# 메모리에 로딩된 값을 api_key 변수에 대입
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
# print(GEMINI_API_KEY)
```

```python
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO

import PIL.Image

image = PIL.Image.open('./imgs/lady.jpg')

client = genai.Client(api_key=GEMINI_API_KEY)

text_input = ('귀여운 포메라니안을 추가해줘',)

response = client.models.generate_content(
    model="gemini-2.0-flash-preview-image-generation",
    contents=[text_input, image],
    config=types.GenerateContentConfig(
      response_modalities=['TEXT', 'IMAGE']
    )
)

for part in response.candidates[0].content.parts:
  if part.text is not None:
    print(part.text)
  elif part.inline_data is not None:
    image = Image.open(BytesIO(part.inline_data.data))
    image.show()
```

## OCR 기술에 대한 설명

### 1. OCR이란? (Optical Character Recognition, 광학 문자 인식)

OCR(Optical Character Recognition, 광학 문자 인식)은 이미지, 스캔 문서, PDF 파일 등에 포함된 텍스트를 컴퓨터가 인식하고 편집할 수 있는 디지털 텍스트로 변환하는 기술입니다. 쉽게 말해, 이미지 속 글자를 텍스트 파일처럼 바꿀 수 있게 해주는 기술입니다.

**작동 원리:**

1.  **이미지 획득:** 스캐너, 카메라 등을 통해 이미지나 문서를 입력받습니다.
2.  **전처리:** 이미지 품질 향상을 위해 노이즈 제거, 기울기 보정, 이진화 등의 과정을 거칩니다.
3.  **문자 영역 분할:** 이미지에서 텍스트 영역을 분리하고, 각 문자를 개별적으로 분할합니다.
4.  **문자 인식:** 분할된 각 문자의 특징을 추출하여 미리 학습된 데이터베이스와 비교하여 어떤 문자인지 판별합니다. 이 단계에서 머신러닝, 특히 딥러닝 기술이 많이 활용됩니다.
5.  **후처리:** 인식된 문자열의 오류를 수정하고, 문맥에 맞게 교정하는 과정을 거칩니다.

### 2. OCR 활용 분야

OCR 기술은 다양한 분야에서 활용되고 있습니다.

*   **문서 관리:** 종이 문서를 디지털화하여 보관하고 검색 효율성을 높입니다. (예: 계약서, 청구서, 영수증 스캔 후 데이터베이스 저장)
*   **자동 데이터 입력:** 서류나 이미지에서 필요한 정보를 자동으로 추출하여 데이터베이스에 입력합니다. (예: 보험 청구서 자동 처리, 은행 수표 자동 판독)
*   **정보 접근성 향상:** 시각 장애인들이 인쇄물을 읽을 수 있도록 텍스트로 변환해줍니다.
*   **번역:** 이미지 속 텍스트를 인식하여 번역합니다.
*   **차량 번호판 인식 (ANPR):** 주차 관리, 교통 감시 등에 활용됩니다.
*   **로봇 자동화 (RPA):** OCR을 이용하여 로봇이 문서 기반의 업무를 자동화할 수 있습니다.
*   **광고 및 마케팅:** 이미지 속 텍스트 정보를 활용하여 광고 효과를 측정하거나, 특정 키워드에 대한 광고를 노출합니다.
*   **교육:** 학생들의 필기 노트를 디지털화하거나, 교재의 텍스트를 추출하여 활용합니다.

### 3. OCR 활용 라이브러리

다양한 OCR 라이브러리가 존재하며, 각 라이브러리는 특징과 장단점을 가지고 있습니다.

*   **Tesseract OCR:** 가장 널리 사용되는 오픈소스 OCR 엔진입니다. 다양한 언어를 지원하며, 비교적 정확도가 높습니다. (Python, C++, Java 등 다양한 언어에서 사용 가능)
*   **Google Cloud Vision API OCR:** Google의 클라우드 기반 OCR 서비스입니다. 강력한 성능과 다양한 기능을 제공하지만, 유료입니다.
*   **Microsoft Azure Computer Vision OCR:** Microsoft의 클라우드 기반 OCR 서비스입니다. Google Cloud Vision API와 유사한 기능을 제공하며, 역시 유료입니다.
*   **Amazon Textract:** Amazon의 클라우드 기반 OCR 서비스입니다. 특히 문서 분석에 특화되어 있습니다.
*   **Keras-OCR:** 딥러닝 기반의 OCR 라이브러리입니다. Tesseract OCR보다 정확도가 높지만, 학습 데이터가 필요합니다.
*   **EasyOCR:** 사용하기 쉬운 OCR 라이브러리입니다. 간단한 이미지의 텍스트를 인식하는 데 적합합니다.
*   **PaddleOCR:** 딥러닝 프레임워크 PaddlePaddle 기반의 OCR 라이브러리입니다. 다양한 언어를 지원하며, 상업적인 용도로도 사용하기에 적합합니다.

**Python에서 많이 사용되는 라이브러리:**

*   **pytesseract:** Tesseract OCR 엔진을 Python에서 사용할 수 있도록 래핑한 라이브러리입니다.
*   **opencv (cv2):** 이미지 처리 라이브러리. OCR 전처리 단계에서 이미지 품질을 개선하는 데 사용됩니다.

### 4. 성능이 가장 좋은 OCR 서비스

"성능이 가장 좋은" OCR 서비스는 상황에 따라 달라질 수 있습니다. 어떤 종류의 문서(예: 깨끗한 인쇄물 vs. 손글씨, 영수증, 표)를 처리해야 하는지, 어떤 언어를 지원해야 하는지, 예산은 얼마인지 등에 따라 최적의 선택이 달라집니다.

**일반적으로 높은 평가를 받는 OCR 서비스:**

*   **Google Cloud Vision API OCR:** 뛰어난 정확도와 다양한 기능을 제공합니다. 특히 복잡한 문서 구조나 손글씨 인식에서 강점을 보입니다. 하지만 사용량에 따라 비용이 발생합니다.
*   **Microsoft Azure Computer Vision OCR:** Google Cloud Vision API와 유사한 수준의 성능을 제공하며, 가격 정책이나 다른 Azure 서비스와의 연동 측면에서 장점이 있을 수 있습니다.
*   **Amazon Textract:** 문서 분석에 특화된 서비스로, 표나 양식에서 데이터를 추출하는 데 뛰어난 성능을 보입니다.
*   **ABBYY FineReader:** 오랜 역사를 가진 상용 OCR 소프트웨어입니다. 로컬 환경에서 작동하며, 강력한 기능과 정확도를 제공합니다.

**무료 또는 오픈 소스 OCR 서비스:**

*   **Tesseract OCR:** 오픈 소스이기 때문에 무료로 사용할 수 있으며, 지속적으로 성능이 개선되고 있습니다. 하지만 상용 서비스에 비해 정확도가 떨어질 수 있습니다. 간단한 문서나 명확한 텍스트에는 충분히 좋은 성능을 보입니다.

**결론:**

가장 좋은 OCR 서비스를 선택하기 위해서는 **테스트**가 중요합니다.  무료 평가판이나 오픈 소스 라이브러리를 사용하여 실제 문서에 대한 성능을 직접 비교해보고, 필요한 기능과 예산을 고려하여 최적의 솔루션을 선택하는 것이 좋습니다.
