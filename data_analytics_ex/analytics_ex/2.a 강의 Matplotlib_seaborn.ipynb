```python
sb.color_palette()
```

```python
base_color = sb.color_palette()[1]
base_color
```

```python
sb.countplot(df, x='generation_id', color=base_color)
plt.title("counts of generation")
plt.show()
```

```python
base_order = df['generation_id'].value_counts().index
```

```python
sb.countplot(data = df, x = 'generation_id' , color= base_color , order= base_order )
plt.show()
```

```python
df['type_1'].value_counts()
```

```python
base_order=df['type_1'].value_counts().index
```

```python
sb.countplot(data=df, x='type_1', order=base_order, color=base_color)
plt.title('# of type_1')
plt.xticks(rotation=90)
plt.show()
```

```python
sb.countplot(data=df, y='type_1', order=base_order, color=base_color)
plt.show()
```

# Pie Charts

```python
df
```

```python
data = df['generation_id'].value_counts()
data
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

plt.pie(
    data,  # generation_id
    labels=data.index,
    startangle = 90,  # 시작 위치 각도
    wedgeprops = {'width': 0.7}
)
plt.title("게임 세대별 갯수")
plt.legend(loc="lower right") 
plt.show()
```

## Heat Maps

```python
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

plt.hist2d(data=df, x='displ', y='comb', cmin=0.5, cmap='viridis_r', bins=20)
plt.colorbar()
plt.title('배기량과 연비의 관계')
plt.xlabel('배기량')
plt.ylabel('연비')
plt.show()
```

## 문제2. 엔진 크기와 이산화탄소 배출량의 관계를 분석하세요.

```python
plt.hist2d(data=df, x='displ', y='co2', cmin=0.5, cmap='viridis_r', bins=20)
plt.colorbar()
plt.title('엔진 크기와 연비의 관계')
plt.xlabel('displ')
plt.ylabel('co2')
plt.show()
```

```python
df.columns
```

```python
corr=df[['displ', 'co2']].corr(numeric_only=True)
sb.heatmap(corr,
    vmin=0.86,
    vmax=1.00,
    cbar=True,
    linewidths=0,
    annot=True,
    cmap='coolwarm'
)

plt.title('엔진 크기와 이산화탄소 배출량의 관계')
plt.show()
```
