## 문제5

```python
# 별점 시스템을 만들려고 한다. 따라서 별점에 대한 데이터가 필요하다. 
# 별점은 1,2,3 즉 3개로 만들것이다. 포인트가 95점 이상이면 3점, 85점 이상이면 2점, 나머지는 1점으로 할 것이다. 
# 리뷰데이터를 통해 각 데이터의 별점을 구하시오.
```

```python
df.columns
```

```python
# 별점 시스템을 만들려고 한다. 따라서 별점에 대한 데이터가 필요하다. 
# 포인트가 95점 이상이면 3점
# 85점 이상이면 2점
# 나머지는 1점으로 할 것이다. 

# points >= 95 -> 3점
# points >= 85 -> 2점
# else(points < 85) -> 1점

# df['points']으로 별점 판단
# 별점 판단하는 함수 정의 후 apply 함수 적용
# 1. 함수 정의, 테스트
```

```python
# 정의부
def check_star(points):
    if points >= 95:
        return "***"
    elif points >= 85:
        return "**"
    else:
        return "*"
```

```python
# 호출부
# 하나의 값(스칼라 값)으로 테스트
point = 60
check_star(point)
```

```python
# 2. apply()에 적용하고
# 컨테이너자료형.apply()
df['points'].apply(check_star)
```

```python
# 3. df에 적용, star_point 컬럼 생성

df['star_point'] = df['points'].apply(check_star)
df
```

```python
# 별3개짜리도 잘되는지 확인
df.loc[df['points']>=95, ['points','star_point']]
```
